# 1. Autoregressive and Multiple Linear Regressive Model

```python
def linearRegress(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results

def regressionPredict(x,res):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    return res.predict(X)

def linearRegressionModel(candidatesList,dfMasterTrain,dfMasterTest,p_vector):
    
    xArrTrain, yArrTrain ,xArrTest = shiftData(candidatesList,dfMasterTrain,dfMasterTest,p_vector)
    #train    
    result = linearRegress(yArrTrain, xArrTrain)
    #predict
    yPred = regressionPredict(xArrTest,result)
    
    return result,yPred
```

#3. ARMA Model

```python
def armaModel(data, ar, ma):
        
    trainingRatio = 0.6
    trainSize = np.floor(len(data) * trainingRatio) #60:40 ratio

    timeSeries = np.array(data)
    timeSeries=timeSeries[::-1]
    
    #plotAcfPacf(timeSeries)
    
    arma_mod20 = sm.tsa.ARMA(timeSeries,(ar,ma), dates=None).fit()
    #print(arma_mod20.params)
    yPredARMA= arma_mod20.fittedvalues[np.int(trainSize):]
    y = timeSeries[np.int(trainSize):]
    return yPredARMA, y

def plotAcfPacf(timeSeries):
    
    acf_values = sm.tsa.stattools.acf(timeSeries, unbiased=False, nlags=10)
    pacf_values = sm.tsa.stattools.pacf(timeSeries, nlags=10)

    #Plot ARMA autocorrelation and partial autocorrelation factors
    fig = plt.figure(figsize=(18,12))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(timeSeries, lags=50, ax=ax1)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(timeSeries, lags=50, ax=ax2)
    #plt.savefig('ARMA_ACF_PACF.png')
```

##Compute Errors

```python
def computeCorrectedSD(newPredy,yActual):
    return round(np.std((newPredy-yActual)/yActual),4)

def computeProbability(newPredy,yActual, newSD):
    pos = 0
    for i in range(0,len(newPredy)):
        if(np.absolute((newPredy[i]-yActual[i])/yActual[i]) <= np.absolute(newSD)):
            pos += 1
    conf = round((1. * pos/len(newPredy) ),4)
    return conf

def computeErrors(yPred, yArrTest):

    #Error in prediction
    errPred = yPred - yArrTest
    #Mean Error
    avgErr=np.mean(errPred)
    #Subtracting the mean error
    yPred = yPred - avgErr
    #Updated error in prediction
    errPred = yPred - yArrTest
    
    #Relative Error
    errRel= 100 * (np.absolute(errPred) / yArrTest)
    meanErrRel = np.mean(errRel)
    meanErrRel = round(meanErrRel,2)
    #RMSE
    errRMSE = np.sqrt(mean_squared_error(yArrTest,yPred))
    errRMSE = round(errRMSE,2)
    #Absolute Error
    errAbs= np.absolute(yPred-yArrTest)
    meanErrAbs = np.mean(errAbs)
    meanErrAbs = round(meanErrAbs,2)
    
    errPlot = errPred/yArrTest
    
    newSD = computeCorrectedSD(yPred,yArrTest)
    prob = computeProbability(yPred,yArrTest,newSD)
    
    return meanErrRel,meanErrAbs,errRMSE,errPlot,newSD,prob
```

##Plot Error Histogram

```python
def plotErrorHistogram(error,nBins,modelName):
    fig, axes = ppl.subplots(1, 1, figsize=(18,12))
    axes.set_xlim(-0.4, 0.4)
    axes.set_ylim(0, 50)
    axes.set_ylabel('Error Frequency')
    axes.set_xlabel('Error')
    axes.set_title(modelName)

    #sb.distplot(error, kde=False, fit=stats.norm, bins=nBins);
    ppl.hist(error, bins=nBins, grid='y', label=modelName)
    plt.savefig('images/'+modelName+'.png')
```

### Generate Model Name

```python
def generateModelName(model, modelVector, candidatesList):
    modelName = model +' Model - Parameters: '
    for i in range(len(modelVector)):
        if(modelVector[i] > 0):
            modelName += (candidatesList[i][:-6] +' ')
    return modelName
```

##Predict Future Prices

```python
def PredictNextMonth(p_vector, candidatesList, modelRes, dfTest):
    xArrTest = []
    n = len(p_vector)
    for i in range(0,n,1):
        k = p_vector[i]
        for j in range(0,k,1):
            xArrTest.append(np.array(dfTest[candidatesList[i]]))
    
    #shift
    curr=0
    for it in range(0,n,1):
        k = p_vector[it]
        for i in range(0,k,1):
            l=len(xArrTest[curr])
            for j in range(0,l,1):
                if(j<l-i-1):
                    xArrTest[curr][j]=xArrTest[curr][j+i+1]
                else:
                    xArrTest[curr][j]=xArrTest[curr][l-1]
            curr=curr+1
            
    return regressionPredict(xArrTest,modelRes)
```

```python
def computePrediction(modelRes, dfMasterTrain, dfMasterTest, modelVector, yPred, candidatesList):
    
    cols = np.array(candidatesList)
    cols = np.insert(cols,0,'Date')
    
    dfPred = DataFrame(data=dfMasterTest[:1], columns=cols)
    dfPred = dfPred.append(dfMasterTest,ignore_index=True)
    errRel = 100. * np.absolute((yPred - dfMasterTest['Oil_Value'])) / dfMasterTest['Oil_Value']
    
    #Predict Current Month
    yPredNew = PredictNextMonth(modelVector,candidatesList,modelRes,dfPred)
    dfPred.loc[0,'Oil_Value']=yPredNew[0]
    
    #Predict Next Month
    dfPredNext = DataFrame(data=dfPred[:1], columns=cols)
    dfPredNext = dfPredNext.append(dfPred,ignore_index=True)
    yPredNew = PredictNextMonth(modelVector,candidatesList,modelRes,dfPredNext)
    dfPredNext.loc[0,'Oil_Value']=yPredNew[0]
    dfPredNext.loc[0,'Date'] = '12/30/2014'
    dfPredNext.loc[1,'Date'] = '11/30/2014'
    
    #Compute Confidence
    a = len(errRel[errRel <= np.mean(errRel)])
    b = len(errRel)
    #conf = round((1. * a/b ),2) 
    predVal = round(dfPred.loc[0,'Oil_Value'],2)
    #meanRelErr = round(np.mean(errRel),2)
    #strResult = "\nThe predicted value is "+str(predVal) + " +/- " \
                    #+ str(meanRelErr)+"% with " + str() + "% confidence \n"
    
    return predVal
```

## Setting up data and running models

```python
def runAutoregressionModel(commodity, inputModels, model, maxMonths, plotHistogram):
    
    if commodity == 'Oil':
        dfMaster,candidatesList = generateOilDfMaster()
    elif commodity == 'Gold':
        dfMaster,candidatesList = generateGoldDfMaster()
    else:
        print("Check the 'commodity' parameter!")
        return None
    
    #Split data into training and testing data
    trainingRatio = 0.6
    trainSize = np.floor(len(dfMaster['Date']) * trainingRatio) #60:40 ratio
    dfMasterTrain = dfMaster[len(dfMaster)-np.int(trainSize):len(dfMaster)]
    dfMasterTest = dfMaster[0:(len(dfMaster)-np.int(trainSize))-1]
    #Result Data Frame
    dfResult = DataFrame(data=None, columns=['Model', 'Months', 'Relative_Error(%)', 'Mean_Absolute_Error', \
                                         'Mean_RMS_Error']) #,'SD','Probability','Prediction'
    
    ctr = 0
    relErrModelMaster = []
    
    #For each model in inputModels,
    for m in range(len(inputModels)):
        
        modelVector = inputModels[m]
        nMonths = []
        relErrModel = []
        
        #Run each model for the number of months in range 1 to maxMonths
        for nMonth in range(1,maxMonths+1):
        
            if(nMonth > 1):
                modelVector[modelVector == (nMonth-1)] = nMonth
            
            yArrTest = np.array(dfMasterTest[candidatesList[0]])
  
            #Run the model, get errors
            modelRes,yPred = linearRegressionModel(candidatesList,dfMasterTrain,dfMasterTest,modelVector)
            meanErrRel,meanErrAbs,meanErrRMSE,errPlot,sd,prob = computeErrors(yPred, yArrTest)
            
            #Non-Linear
            #modelRes,yPred = nonLinearRegressiveModel(dfMasterTrain,dfMasterTest,modelVector)
            #modelName = generateModelName('Non-Linear Regression', modelVector, candidatesList)
            
            relErrModel.append(meanErrRel)
            
            nMonths.append(nMonth)
        
            #Predict next month
            predVal = computePrediction(modelRes, dfMasterTrain, dfMasterTest, modelVector, yPred, candidatesList)
        
            modelName = generateModelName('Autoregressive', modelVector, candidatesList)

            if(nMonth == 3):
                dfResult.loc[ctr,'Model'] = modelName
                dfResult.loc[ctr,'Months'] = nMonth
                dfResult.loc[ctr,'Relative_Error(%)'] = meanErrRel
                dfResult.loc[ctr,'Mean_Absolute_Error'] = meanErrAbs
                dfResult.loc[ctr,'Mean_RMS_Error'] = meanErrRMSE
                #dfResult.loc[ctr,'SD'] = sd
                #dfResult.loc[ctr,'Probability'] = prob
                #dfResult.loc[ctr,'Prediction'] = predVal
                
                if(plotHistogram == True):
                    plotErrorHistogram(errPlot,8 ,commodity + " - " + modelName)
                ctr = ctr + 1

        #ppl.plot(nMonths,relErrModel, linewidth=2, linestyle=linestyles[m-1], color=colors[m-1], label=modelName, marker='o')

        if m==0:
            relErrModelMaster = relErrModel
            modelNameMaster = [modelName]
        if m>0:
            relErrModelMaster = np.vstack((relErrModelMaster,relErrModel))
            modelNameMaster = np.append(modelNameMaster,modelName)
            
    return dfResult, modelNameMaster, relErrModelMaster
```

```python
def runArmaModel(commodity, ar, ma, plotHistogram):
    
    if(commodity=='Oil'):
        data = dfOil['Oil_Value']
    elif(commodity=='Gold'):
        data = dfGold['Gold_Value']
        
    dfResult = DataFrame(data=None, columns=['Model', 'Months', 'Relative_Error(%)', 'Mean_Absolute_Error', \
                                         'Mean_RMS_Error']) #,'SD','Probability','Prediction' 
    relErrArma=[]
    ctr=0
    
    for nMonth in range(1,ar+1):
    
        #Run ARMA Model
        yPredARMA,y=armaModel(data,nMonth,0)
    
        #Compute ARMA Errors
        meanErrRelARMA,meanErrAbsARMA,meanErrRMSEARMA,errPlotARMA, sdARMA, probARMA = computeErrors(yPredARMA, y)
        relErrArma.append(meanErrRelARMA)
        
        #predVal = computePrediction(modelRes, dfMasterTrain, dfMasterTest, modelVector, yPred, candidatesList)
        
        modelName = "Autoregressive Moving Average Model" 
        if(nMonth == 3):
            dfResult.loc[ctr,'Model'] = modelName + "(" + str(3) + ", "+str(ma) + ")"
            dfResult.loc[ctr,'Months'] = nMonth
            dfResult.loc[ctr,'Relative_Error(%)'] = meanErrRelARMA
            dfResult.loc[ctr,'Mean_Absolute_Error'] = meanErrAbsARMA
            dfResult.loc[ctr,'Mean_RMS_Error'] = meanErrRMSEARMA
            #dfResult.loc[ctr,'SD'] = sdARMA
            #dfResult.loc[ctr,'Probability'] = probARMA
            #dfResult.loc[ctr,'Prediction'] = ""
    
            ctr = ctr + 1
            if(plotHistogram == True):
                plotErrorHistogram(errPlotARMA,8 ,modelName)
                
    return dfResult, modelName, relErrArma
```
