# 1. Autoregressive and Multiple Linear Regressive Model

```python
def linearRegress(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results

def regressionPredict(x,res):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    return res.predict(X)

def linearRegressionModel(dfMasterTrain,dfMasterTest,p_vector):
    
    xArrTrain, yArrTrain ,xArrTest = shiftData(dfMasterTrain,dfMasterTest,p_vector)
    #train    
    result = linearRegress(yArrTrain, xArrTrain)
    #predict
    yPred = regressionPredict(xArrTest,result)
    
    return result,yPred
```

# 2. Non Linear Regressive Model

```python
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def nonLinearRegress(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    
    #Polynomial Regression
    model = make_pipeline(PolynomialFeatures(2), Ridge())
    results=model.fit(X, y)
     
    # SVM Regression    
    #svr_rbf = SVR(kernel='poly')
    #results_rbf = svr_rbf.fit(X, y)
    return results

def nonLinearPredict(x,res):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    return res.predict(X)

def nonLinearRegressiveModel(dfMasterTrain,dfMasterTest,p_vector):
    xArrTrain, yArrTrain ,xArrTest = shiftData(dfMasterTrain,dfMasterTest,p_vector)
     #train    
    result = nonLinearRegress(yArrTrain, xArrTrain)
    #predict
    yPred = regressionPredict(xArrTest,result)
   
    return result,yPred
```

##Compute Errors

```python
def computeErrors(yPred, yArrTest):

    #Error in prediction
    errPred = yPred - yArrTest
    #Mean Error
    avgErr=np.mean(errPred)
    #Subtracting the mean error
    yPred = yPred - avgErr
    #Updated error in prediction
    errPred = yPred - yArrTest
    
    #Relative Error
    errRel= 100 * (np.absolute(errPred) / yArrTest)
    meanErrRel = np.mean(errRel)
    meanErrRel = round(meanErrRel,2)
    #RMSE
    errRMSE = np.sqrt(mean_squared_error(yArrTest,yPred))
    errRMSE = round(errRMSE,2)
    #Absolute Error
    errAbs= np.absolute(yPred-yArrTest)
    meanErrAbs = np.mean(errAbs)
    meanErrAbs = round(meanErrAbs,2)
    
    errPlot = errPred/yArrTest
    ''' 
    print ("Mean Relative Error: ", meanErrRel)
    print ("Mean Absolute Error: ", meanAbsRel)
    print ("RMSE: ", errRMSE)
    plotErrorHistogram (errPlot,10,'Error Histogram')
    '''
    return meanErrRel,meanErrAbs,errRMSE,errPlot
```

### Generate Model Name

```python
def generateModelName(model, modelVector, candidatesList):
    modelName = model +' model with factors: '
    for i in range(len(modelVector)):
        if(modelVector[i] > 0):
            modelName += (candidatesList[i][:-6] +' ')
    return modelName
```

##Predict Future Prices

```python
def PredictNextMonth(p_vector, modelRes,dfTest):
    xArrTest = []
    n = len(p_vector)
    for i in range(0,n,1):
        k = p_vector[i]
        for j in range(0,k,1):
            xArrTest.append(np.array(dfTest[candidatesList[i]]))
    
    #shift
    curr=0
    for it in range(0,n,1):
        k = p_vector[it]
        for i in range(0,k,1):
            l=len(xArrTest[curr])
            for j in range(0,l,1):
                if(j<l-i-1):
                    xArrTest[curr][j]=xArrTest[curr][j+i+1]
                else:
                    xArrTest[curr][j]=xArrTest[curr][l-1]
            curr=curr+1
    return regressionPredict(xArrTest,modelRes)

```

```python
def computePrediction(modelRes, dfMasterTrain, dfMasterTest, modelVector, yPred, candidatesList):
    
    cols = np.array(candidatesList)
    cols = np.insert(cols,0,'Date')
    
    dfPred = DataFrame(data=dfMasterTest[:1], columns=cols)
    dfPred = dfPred.append(dfMasterTest,ignore_index=True)
    errRel = 100. * np.absolute((yPred - dfMasterTest['Oil_Value'])) / dfMasterTest['Oil_Value']
    
    #Predict Current Month
    yPredNew = PredictNextMonth(modelVector,modelRes,dfPred)
    dfPred.loc[0,'Oil_Value']=yPredNew[0]
    
    #Predict Next Month
    dfPredNext = DataFrame(data=dfPred[:1], columns=cols)
    dfPredNext = dfPredNext.append(dfPred,ignore_index=True)
    yPredNew = PredictNextMonth(modelVector,modelRes,dfPredNext)
    dfPredNext.loc[0,'Oil_Value']=yPredNew[0]
    dfPredNext.loc[0,'Date'] = '12/30/2014'
    dfPredNext.loc[1,'Date'] = '11/30/2014'
    
    #Compute Confidence
    a = len(errRel[errRel <= np.mean(errRel)])
    b = len(errRel)
    conf = round((1. * a/b ),2) 
    predVal = round(dfPred.loc[0,'Oil_Value'],2)
    meanRelErr = round(np.mean(errRel),2)
    strResult = "\nThe predicted value is "+str(predVal) + " +/- " \
                    + str(meanRelErr)+"% with " + str() + "% confidence \n"
    
    return predVal,meanRelErr,conf 
```
